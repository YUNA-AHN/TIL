```python
def get_deposit_products():
    # 본인의 API KEY 로 수정합니다.
    api_key = "37495baeddeb83f0e69f7e484fc7a445"
    url = 'http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json'
    params = {
        'auth' : api_key,
        'topFinGrpNo' : '020000',
        'pageNo' : 1
    }

    # 요구사항에 맞도록 이곳의 코드를 수정합니다.
    response = requests.get(url,params).json()['result']

    keys  = ['fin_prdt_cd', 'intr_rate', 'save_trm', 'intr_rate_type', 'intr_rate_type_nm', 'intr_rate2']
    kor_keys = ['금융상품코드', '저축 금리', '저축 기간', '저축금리유형', '저축금리유형명', '최고 우대 금리']

    option_list = []
    for info in response['optionList']:
        new_list = {}
        for num in range(len(keys)):
            new_list[kor_keys[num]] = info[keys[num]]
            option_list.append(new_list)

    result = []
    for product in response['baseList']:
        rate_info = []
        fin_prdt_nm = product['fin_prdt_nm']
        kor_co_nm = product['kor_co_nm']
        for option in option_list:
            if product['fin_prdt_cd'] == option['금융상품코드']:
                option_dict = {
                    '저축 금리': option['저축 금리'],
                    '저축 기간': option['저축 기간'],
                    '저축금리유형': option['저축금리유형'],
                    '저축금리유형명': option['저축금리유형명'],
                    '최고 우대 금리': option['최고 우대 금리']
                }
                rate_info.append(option_dict)
        prod_dict = {
            '금리정보' : rate_info,
            '금융상품명' : fin_prdt_nm,
            '금융회사명' : kor_co_nm
            }
        result.append(prod_dict)

    return result
```
